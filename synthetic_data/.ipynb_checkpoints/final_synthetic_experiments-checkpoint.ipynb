```python
# RUN ALGORITHM 1

exercise_1  = {}
max_num_cl = 50 # maximum number of value k
num_boots = 100 # number of bootstrap runs
methods_list = ['AffinityPropagation', 'Kmeans', 'Birch', 'AgglomerativeClustering', 'MeanShift'] 

import time

st = time.time()
ss = StandardScaler()
eps_ls = np.linspace(.1, 2, 3)
for dataset in ['spiral', 'd31', 'r15', 'aggregation']:
    
    print('Starting dataset ', dataset)
    print('\nElapsed time : ', time.time()-st)
    
    exercise_1[dataset] = {}
    data = np.loadtxt('Data/Shapes/'+dataset+'.txt')
    train_set, test_set = random_splitting(data, .5)
    
    for method in methods_list:
        exercise_1[key][method] = {}
        if method == 'AffinityPropagation':
            ari, ami, true_labels, predicted_labels, k_pred, k_true = run_exercise_1(train_set, test_set, method, num_boots, max_num_cl, AFFINITY, LINKAGE)
            exercise_1[key][method]['k_pred'], exercise_1[key][method]['k_true'] = k_pred, k_true            
        else:
            ari, ami, true_labels, predicted_labels = run_exercise_1(train_set, test_set, method, num_boots, max_num_cl, AFFINITY, LINKAGE)
        exercise_1[key][method]['ari'], exercise_1[key][method]['ami'] = ari, ami
        exercise_1[key][method]['true_labels'], exercise_1[key][method]['predicted_labels'] = true_labels, predicted_labels
    
    
    if dataset == 'spiral':
        exercise_1[dataset]['DBSCAN'] = {}
        eps_ls = np.linspace(.3,1.5,3)
        X, Y = ss.fit_transform(train_set), ss.fit_transform(test_set)
        k_train, k_test = np.zeros([len(eps_ls), num_boots]), np.zeros([len(eps_ls), num_boots])
        true_labels, predicted_labels = np.zeros([len(eps_ls), num_boots, len(Y)]), np.zeros([len(eps_ls), num_boots, len(Y)])
        ari, ami = np.zeros([len(eps_ls), num_boots]), np.zeros([len(eps_ls), num_boots])
        for e_,eps in enumerate(eps_ls):
            for b in range(num_boots):
                X_, Y_ =  subsample(X, int(len(X)/3)), subsample(Y, int(len(Y)/3))
                fit_tr = DBSCAN(eps=eps, min_samples=1).fit_predict(X)
                nn_inds = np.argmin(sp_dist.cdist(Y, X, metric = 'euclidean'), axis = 1)
                predicted_labels_ = np.asarray([fit_tr[s] for s in nn_inds])
                true_labels_ = DBSCAN(eps=eps, min_samples=1).fit_predict(Y)
                k_train[e_, b], k_test[e_,b] = len(np.unique(fit_tr)), len(np.unique(true_labels))
                ari[e_,b], ami[e_, b] = sklearn.metrics.adjusted_rand_score(predicted_labels_, true_labels_), sklearn.metrics.adjusted_mutual_info_score(predicted_labels_, true_labels_)
                true_labels[e_, b], predicted_labels[e_, b] = true_labels_, predicted_labels_
        exercise_1['spiral']['DBSCAN']['ari'], exercise_1['spiral']['DBSCAN']['ami'] = ari, ami
        exercise_1['spiral']['DBSCAN']['true_labels'], exercise_1['spiral']['DBSCAN']['predicted_labels'] = true_labels, predicted_labels
    np.save('Results/exercise_1', exercise_1)
    
print('Total elapsed time: ', time.time()-st)
```
