```r
PAM50_LS = sort(c("BIRC5","CDC6","CCNB1","TYMS","MYBL2","CEP55","MELK","NDC80","RRM2","UBE2C","CENPF","PTTG1","EXO1","ORC6L","CCNE1","CDC20","MKI67","KIF2C","ACTR3B","MYC","EGFR","KRT5","PHGDH","CDH3","MIA","KRT17","FOXC1","SFRP1","KRT14","ESR1","SLC39A6","BAG1","MAPT","PGR","MLPH","BCL2","MDM2","NAT1","FOXA1","BLVRA","MMP11","FGFR4","GRB7","ERBB2"))
THREE_GENES_ls = sort(c("ESR1", "ERBB2", "AURKA"))
```

```r
data(mainz)
l_mainz = list()
l_mainz_3 = list()
#mainz_exprs_3 = matrix(0, nrow = 3, ncol = dim(mainz)[2]) 
```

```r
# MAINZ
data(mainz)


for (i in 1:length(PAM50_LS)){
  key = PAM50_LS[i]
  index = grep(paste("\\b",key,"\\b", sep = ""), featureData(mainz)@data[,3])
  print(key)
  print(index)
  if (length(index) == 1){
    l_mainz[key] = index
  }
  if (length(index) > 1){
    l_mainz[key] = index[1]
  }
  if (length(index) == 0){
    l_mainz[key] = 0
  }
}
```

```r
# TRANSBIG

data("transbig")
l_tra = list()
l_tra_3 = list()

for (i in 1:length(PAM50_LS)){
  key = PAM50_LS[i]
  index = grep(paste("\\b",key,"\\b", sep = ""), featureData(transbig)@data[,3])
  print(key)
  print(index)
  if (length(index) == 1){
    l_tra[key] = index
  }
  if (length(index) > 1){
    l_tra[key] = index[1]
  }
  if (length(index) == 0){
    l_tra[key] = 0
  }
}

tra_exprs_3 = matrix(0, nrow = 3, ncol = dim(transbig)[2]) 

for (i in 1:length(THREE_GENES_ls)){
  key = THREE_GENES_ls[i]
  index = grep(paste("\\b",key,"\\b", sep = ""), featureData(transbig)@data[,3])
  print(length(index))
  for (j in 1:length(index)){
    print(index[j])
    tra_exprs_3[i,] = tra_exprs_3[i,] + exprs(transbig)[index[j],]/length(index)
  }
}

write.csv(tra_exprs_3, "tra_3_AVG") 

for (i in 1:length(THREE_GENES_ls)){
  key = THREE_GENES_ls[i]
  index = grep(paste("\\b",key,"\\b", sep = ""), featureData(transbig)@data[,3])
  print(key)
  print(index)
  if (length(index) == 1){
    l_tra_3[key] = index
  }
  if (length(index) > 1){
    l_tra_3[key] = index[1]
  }
  if (length(index) == 0){
    l_tra_3[key] = 0
  }
}

l_tra_act = which(l_tra != 0)
exprs_tra_pam = exprs(transbig)[l_tra_act,]

l_tra_act_3 = which(l_tra_3 != 0)
exprs_tra_3 = exprs(transbig)[l_tra_act_3,]


# UNT 
data("unt")

l_unt = list()
l_unt_3 = list()

for (i in 1:length(PAM50_LS)){
  key = PAM50_LS[i]
  index = grep(paste("\\b",key,"\\b", sep = ""), featureData(unt)@data[,3])
  print(key)
  print(index)
  if (length(index) == 1){
    l_unt[key] = index
  }
  if (length(index) > 1){
    l_unt[key] = index[1]
  }
  if (length(index) == 0){
    l_unt[key] = 0
  }
}

unt_exprs_3 = matrix(0, nrow = 3, ncol = dim(unt)[2]) 

for (i in 1:length(THREE_GENES_ls)){
  key = THREE_GENES_ls[i]
  index = grep(paste("\\b",key,"\\b", sep = ""), featureData(transbig)@data[,3])
  print(length(index))
  for (j in 1:length(index)){
    print(index[j])
    unt_exprs_3[i,] = unt_exprs_3[i,] + exprs(unt)[index[j],]/length(index)
  }
}

write.csv(unt_exprs_3, "unt_3_AVG") 



for (i in 1:length(THREE_GENES_ls)){
  key = THREE_GENES_ls[i]
  index = grep(paste("\\b",key,"\\b", sep = ""), featureData(unt)@data[,3])
  print(key)
  print(index)
  if (length(index) == 1){
    l_unt_3[key] = index
  }
  if (length(index) > 1){
    l_unt_3[key] = index[1]
  }
  if (length(index) == 0){
    l_unt_3[key] = 0
  }
}



l_unt_act = which(l_unt != 0)
exprs_unt_pam = exprs(unt)[l_unt_act,]

l_unt_act_3 = which(l_unt_3 != 0)
exprs_unt_3 = exprs(unt)[l_unt_act_3,]


#VDX
data("vdx")
l_vdx = list()
l_vdx_3 = list()

for (i in 1:length(PAM50_LS)){
  key = PAM50_LS[i]
  index = grep(paste("\\b",key,"\\b", sep = ""), featureData(vdx)@data[,3])
  print(key)
  print(index)
  if (length(index) == 1){
    l_vdx[key] = index
  }
  if (length(index) > 1){
    l_vdx[key] = index[1]
  }
  if (length(index) == 0){
    l_vdx[key] = 0
  }
}

for (i in 1:length(THREE_GENES_ls)){
  key = THREE_GENES_ls[i]
  index = grep(paste("\\b",key,"\\b", sep = ""), featureData(vdx)@data[,3])
  print(key)
  print(index)
  if (length(index) == 1){
    l_vdx_3[key] = index
  }
  if (length(index) > 1){
    l_vdx_3[key] = index[1]
  }
  if (length(index) == 0){
    l_vdx_3[key] = 0
  }
}

l_vdx_act = which(l_vdx != 0)
exprs_vdx_pam = exprs(vdx)[l_vdx_act,]

l_vdx_act_3 = which(l_vdx_3 != 0)
exprs_vdx_3 = exprs(vdx)[l_vdx_act_3,]



vdx_exprs_3 = matrix(0, nrow = 3, ncol = dim(vdx)[2]) 

for (i in 1:length(THREE_GENES_ls)){
  key = THREE_GENES_ls[i]
  index = grep(paste("\\b",key,"\\b", sep = ""), featureData(vdx)@data[,3])
  print(length(index))
  for (j in 1:length(index)){
    print(index[j])
    vdx_exprs_3[i,] = vdx_exprs_3[i,] + exprs(vdx)[index[j],]/length(index)
  }
}

write.csv(vdx_exprs_3, "vdx_3_AVG") 
```
